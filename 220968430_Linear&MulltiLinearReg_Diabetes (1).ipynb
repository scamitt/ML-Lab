```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from mpl_toolkits import mplot3d
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

# Linear Regression

```python
df_incl_textual=pd.read_csv("diabetes.csv")
```

```python
df=df_incl_textual.select_dtypes(include='number')
```

```python
sns.heatmap(df.corr())
```

```python
df.corr()
```

Regression between wt and disp

```python
x = np.array(df['Insulin']).reshape(-1, 1)
y = np.array(df['SkinThickness']).reshape(-1, 1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
x_train,y_train
```

```python
plt.scatter(x_train,y_train)
plt.title("FOR TRAINING DATASET")
```

```python
regr=LinearRegression()
```

```python
regr.fit(x_train,y_train)
```

```python
c=regr.intercept_
m=regr.coef_
```

```python
print(c)
print(m)
```

```python
y_pred=regr.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
plt.scatter(x_test,y_test)

plt.plot(x_test,m*x_test+c, color ='r')
plt.title("PREDICTION RESULTS")
```

```python
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mse)

print(f"Mean Square Error (MSE): {mse}")
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Root Mean Square Error (RMSE): {rmse}")
```
